```python
file = pd.concat([file00, file01, file02, file03, file04, file05, file06, file07, file08, file09])
file
```

```python
file['TS_Central'] = pd.to_datetime(file['Timestamp'],  unit='ms')
file['TS_Central'] = file['TS_Central'].dt.tz_localize('UTC').dt.tz_convert('US/Central')
file
```

```python
file.dtypes
```

```python
#Para pasar en formato date
#file['Timestamp2'] = file['Timestamp1'].values.astype('datetime64[D]')

#Lo queremos en formato str
file['TS_Text'] = file['TS_Central'].astype(str)
file
#file['TS_Text'] = file['TS_Text'].str[:10]
```

```python
file['AvantiID'] = file['Device_ID'] + '_' + file['TS_Text'].str[:10]
file
```

```python
#delte identicals
file_drop = file.drop_duplicates(subset=['Device_ID', 'Lat', 'Lon', 'Timestamp', 'OS'])
file_drop
```

```python
file_drop.to_csv(r'F:\Quadrant\Procesos\DescargasBigQuery\20210618_AcuñaDelRio\DespuesPython\Quadrant_AcuñaDelRio_CT_Clean.csv')
```
