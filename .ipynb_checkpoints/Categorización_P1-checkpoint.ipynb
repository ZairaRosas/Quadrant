```python
import pandas as pd
import geopandas as gpd
from pandas import read_csv
from geopandas import GeoDataFrame, points_from_xy
from shapely import wkt
from geopandas.tools import sjoin
```

```python
#df = read_csv(r'F:\Quadrant\Procesos\DescargasBigQuery\20210618_AcuñaDelRio\DespuesPython\Quadrant_AcuñaDelRio_CT_Clean.csv', low_memory = False)
#df
```

```python
df = gpd.read_file(r'F:\Quadrant\Procesos\SHP\Acuña_DelRio\AcuñaDelRio_1am_6am_sample.shp')
df
```

Quadrant de 1 am a 6 am

```python
df_1am_6am = df[df['TS_Text'].str.contains(' 01:| 02:| 03:| 04:| 05:| 06:') ]
df_1am_6am
```

```python
gdf_1am_6am = GeoDataFrame(df_1am_6am, geometry=points_from_xy(df_1am_6am['Lon'], df_1am_6am['Lat']))
gdf_1am_6am = gdf_1am_6am.set_crs("EPSG:4326")
gdf_1am_6am
```

Quadrant de 3 am para determinar hogar

```python
df_3am = df[df['TS_Text'].str.contains(' 03:') ]
df_3am
```

```python
df_3am_unique = df_3am.drop_duplicates(subset=['Device_ID'])
df_3am_unique
```

```python
gdf_3am_unique = GeoDataFrame(df_3am_unique, geometry=points_from_xy(df_3am_unique['Lon'], df_3am_unique['Lat']))
gdf_3am_unique
```

#### Crear buffer

```python
#20m = 0.00020 
gdf_3am_unique['buff_geom'] = gdf_3am_unique.geometry.buffer(.00020)
gdf_3am_unique
```

```python
del gdf_3am_unique['geometry']
```

```python
#gdf_3am_buff = GeoDataFrame(gdf_3am_unique, geometry = 'buff_geom')
#gdf_3am_buff
```

```python
gdf_3am_buff = gpd.read_file(r'F:\Quadrant\Procesos\SHP\Acuña_DelRio\AcuñaDelRio_buff15_3am_unique.shp')
gdf_3am_buff
```

#### Joins

```python
sjoin = sjoin(gdf_1am_6am, gdf_3am_buff, how="left", op="intersects")
sjoin
```

```python
sjoin_ID = sjoin.loc[sjoin['Device_ID_left'] == sjoin['Device_ID_right']]
sjoin_ID
```

```python
counts_sjoinID = sjoin_ID.groupby(['Device_ID_left']).size().reset_index(name='counts')
counts_sjoinID.sort_values(by='counts')
```

```python
counts_df_1am_6am = df_1am_6am.groupby(['Device_ID']).size().reset_index(name='counts')
counts_df_1am_6am.sort_values(by='counts')
```

```python
indice_porc = pd.merge(counts_sjoinID, counts_df_1am_6am,  how='left', left_on='Device_ID_left', right_on = 'Device_ID')
indice_porc
```

#### Crear índice porcentaje

```python
indice_porc['indice'] = (100 *  indice_porc.counts_x) / indice_porc.counts_y
indice_porc
```

```python
indice_porc70 = indice_porc[indice_porc['indice'] >= 70 ]
indice_porc70
```

```python
id_finals1 = pd.merge(indice_porc70, gdf_3am_unique,  how='left', left_on='Device_ID_left', right_on = 'Device_ID')
id_finals1
```

```python
id_finals = id_finals1[['Device_ID_left','Lon', 'Lat','counts_x', 'counts_y', 'indice']]
id_finals.columns = ['Device_ID','Lon', 'Lat','Tot3am', 'Total', 'Indice']
id_finals
```

```python
gdf_id_finals = GeoDataFrame(id_finals, geometry=points_from_xy(id_finals['Lon'], id_finals['Lat']))
gdf_id_finals = gdf_id_finals.set_crs("EPSG:4326")
gdf_id_finals
```
