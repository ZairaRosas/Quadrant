```python
df = gpd.read_file(r'F:\Quadrant\Procesos\SHP\Acuña_DelRio\AcuñaDelRio_1am_6am_sample.shp')
df
```

Quadrant de 1 am a 6 am

```python
df_1am_6am = df[df['TS_Text'].str.contains(' 01:| 02:| 03:| 04:| 05:| 06:') ]
df_1am_6am
```

```python
gdf_1am_6am = GeoDataFrame(df_1am_6am, geometry=points_from_xy(df_1am_6am['Lon'], df_1am_6am['Lat']))
gdf_1am_6am
```

Quadrant de 3 am para determinar hogar

```python
df_3am = df[df['TS_Text'].str.contains(' 03:') ]
df_3am
```

```python
df_3am_unique = df_3am.drop_duplicates(subset=['Device_ID'])
df_3am_unique
```

```python
gdf_3am_unique = GeoDataFrame(df_3am_unique, geometry=points_from_xy(df_3am_unique['Lon'], df_3am_unique['Lat']))
gdf_3am_unique
```

```python
#20m = 0.00020 
gdf_3am_unique['buff_geom'] = gdf_3am_unique.geometry.buffer(.00020)
gdf_3am_unique
```

```python
del gdf_3am_unique['geometry']
```

```python
gdf_3am_buff = GeoDataFrame(gdf_3am_unique, geometry = 'buff_geom')
gdf_3am_buff
```
