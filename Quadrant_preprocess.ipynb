```python
import pandas as pd
from pandas import read_csv
import geopandas as gpd
from geopandas import GeoDataFrame
from shapely.geometry import Point, LineString, shape
```

```python
file = read_csv(r'E:\Quadrant_Sin_comprimir\02\part-00000-c6da3795-1b1c-4509-bfa7-0de1833d6b1a-c000.csv', names = ['Device_ID', 'Lat', 'Lon', 'GPS_Accu', 'Timestamp', 'IP', 'OS_ID', 'OS', 'Country', 'Quad_ID', 'Publisher'])
file
```

```python
file['Timestamp1'] = file['Timestamp'].astype(str)
file['Timestamp1'] = file['Timestamp1'].str[:10]
file['Timestamp1'] = pd.to_datetime(file['Timestamp1'], unit='s')
file
```

```python
#Para pasar en formato date
file['Timestamp2'] = file['Timestamp1'].values.astype('datetime64[D]')

#Lo queremos en formato str
file['Timestamp2'] = file['Timestamp1'].astype(str)
file['Timestamp2'] = file['Timestamp2'].str[:10]
file
```

```python
file['AvantiID'] = file['Device_ID'] + file['Timestamp2']
file
```

```python
#verificar si el archivo tiene datos del mismo devide

unique_values_Did =  file.AvantiID.value_counts()
unique_values_Did
```

```python
#prueba con el device que mÃ¡s se repite

file_prueba = file[(file['AvantiID'] == '3dd899fb-e89c-4c72-bcd4-aeed177de8c52019-10-02') | (file['AvantiID'] == '6f314ba2-2d59-414e-adf3-52834b4e76de2019-10-01') | (file['AvantiID'] == '7FB3AE1C-FA3A-4E8D-9C44-A6BCF59C1F812019-10-01')  ]
file_prueba.sort_values(by='Timestamp1', ascending=True)
```

```python
#zip de coordinates into a point object and convert to a GeoDataFrame

geometry = [Point(xy) for xy in zip(file_prueba.Lon, file.Lat)]
geo_df = GeoDataFrame(file_prueba, geometry=geometry)
geo_df = geo_df.sort_values(by='Timestamp1', ascending=True)
geo_df
```

```python
geo_df2 = geo_df.groupby(['AvantiID', 'Device_ID', 'OS_ID', 'OS', 'Country', 'Quad_ID', 'Publisher'])['geometry'].apply(lambda x:
                                      LineString(x.tolist()))
geo_df3 = GeoDataFrame(geo_df2, geometry = 'geometry')
geo_df3
```
