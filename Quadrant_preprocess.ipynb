```python
file = pd.concat([file01, file02, file03, file04, file05, file06, file07, file08, file09, file10])
file
```

```python
file['Timestamp1'] = file['Timestamp'].astype(str)
file['Timestamp1'] = file['Timestamp1'].str[:10]
file['Timestamp1'] = pd.to_datetime(file['Timestamp1'], unit='s')
file
```

```python
#Para pasar en formato date
file['Timestamp2'] = file['Timestamp1'].values.astype('datetime64[D]')

#Lo queremos en formato str
file['Timestamp2'] = file['Timestamp1'].astype(str)
file['Timestamp2'] = file['Timestamp2'].str[:10]
file
```

```python
file['AvantiID'] = file['Device_ID'] + file['Timestamp2']
file
```

```python
#delte identicals
file_drop = file.drop_duplicates(subset=['Device_ID', 'Lat', 'Lon', 'Timestamp', 'OS'])
file_drop
```

```python
file_drop.to_csv(r'E:\Procesos\DescargasBigQuery\20210528_Torreon_Matamoros\DespuesPython\Quadrant_Torreon_clean.csv')
```

```python
#zip de coordinates into a point object and convert to a GeoDataFrame

geometry = [Point(xy) for xy in zip(file_prueba.Lon, file.Lat)]
geo_df = GeoDataFrame(file_prueba, geometry=geometry)
geo_df = geo_df.sort_values(by='Timestamp1', ascending=True)
geo_df
```
