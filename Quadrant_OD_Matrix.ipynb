# Proceso para generación de OD Matrix

```python
import pandas as pd
from pandas import read_csv
```

```python
starts = read_csv(r'F:\Quadrant\Procesos\BD\Acuña_DelRio\Quadrant_AcuñaDelRio_Starts_JoinDiv_completo.csv')
ends = read_csv(r'F:\Quadrant\Procesos\BD\Acuña_DelRio\Quadrant_AcuñaDelRio_Ends_JoinDiv_completo.csv')
```

```python
starts
```

```python
del starts['OBJECTID']
del starts['OBJECTID_12']
starts
```

```python
del ends['OBJECTID']
del ends['OBJECTID_12']
ends
```

```python
starts_ends = pd.merge(starts, ends, on = 'AvantiID')
starts_ends
```

```python
starts_ends['z_visum_x'] = starts_ends['z_visum_x'].astype(str)
starts_ends['z_visum_y'] = starts_ends['z_visum_y'].astype(str)
starts_ends 
```

```python
starts_ends['StartsEnds'] = starts_ends['z_visum_x'] + " " + starts_ends['z_visum_y']
starts_ends
```

```python
matrix = starts_ends.StartsEnds.value_counts().rename_axis('StartsEnds').reset_index(name='count')
matrix
```

```python
matrix[['Starts', 'Ends']] = matrix['StartsEnds'].str.split(' ',1, expand =True)
matrix
```

```python
matrix['Counts'] = matrix['count']
matrix
```

```python
del matrix['StartsEnds']
del matrix['count']
matrix
```

```python
matrix.to_csv(r'F:\Quadrant\Procesos\BD\Acuña_DelRio\ODMatrix_AcuñaDelRio_completo.csv')
```
