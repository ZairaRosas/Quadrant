# Proceso para generación de OD Matrix

```python
import pandas as pd
from pandas import read_csv
```

```python
starts = read_csv(r'F:\Quadrant\Procesos\BD\Quadrant_agebJoin_starts.csv')
ends = read_csv(r'F:\Quadrant\Procesos\BD\Quadrant_agebJoin_ends.csv')
```

```python
starts
```

```python
del starts['OBJECTID_12']
del starts['OBJECTID']
del starts['ORIG_FID']
del starts['OBJECTID_1']
del starts['CVE_ENT']
del starts['CVE_MUN']
del starts['CVE_AGEB']
del starts['CVE_LOC']
del starts['PobTot']
starts
```

```python
del ends['OBJECTID_12']
del ends['OBJECTID']
del ends['ORIG_FID']
del ends['OBJECTID_1']
del ends['CVE_ENT']
del ends['CVE_MUN']
del ends['CVE_AGEB']
del ends['CVE_LOC']
del ends['PobTot']
ends
```

```python
starts_ends = pd.merge(starts, ends, on = 'AvantiID')
starts_ends
```

```python
starts_ends['StartsEnds'] = starts_ends['CVEGEO_x'] + " " + starts_ends['CVEGEO_y']
starts_ends
```

```python
matrix = starts_ends.StartsEnds.value_counts().rename_axis('StartsEnds').reset_index(name='count')
matrix
```

```python
matrix[['Starts', 'Ends']] = matrix['StartsEnds'].str.split(' ',1, expand =True)
matrix
```

```python
matrix['Counts'] = matrix['count']
matrix
```

```python
del matrix['StartsEnds']
del matrix['count']
matrix
```

```python
matrix.to_csv(r'F:\Quadrant\Procesos\BD\Torreon\ODMatrix_Torreon_20191013.csv')
```
